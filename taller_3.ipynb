```python
from sklearn.impute import SimpleImputer
columnas_numericas=df.select_dtypes(include=['number']).columns
columnas_numericas
```

```python
imputacion_numerica=SimpleImputer(missing_values=np.nan,strategy='mean')
imputacion_numerica
df[columnas_numericas]=imputacion_numerica.fit_transform(df[columnas_numericas])
df
```

```python
columnas_texto=df.select_dtypes(exclude=['number']).columns
columnas_texto
```

```python
imputacion_texto=SimpleImputer(missing_values=np.nan,strategy='most_frequent')
imputacion_texto
df[columnas_texto]=imputacion_texto.fit_transform(df[columnas_texto])
df
```
